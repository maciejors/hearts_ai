```python
import sys
sys.path.append('..')
```

```python
from hearts_ai.rl.results import (
    load_training_results, 
    PlotMakerPlaying,
)
```

```python
def log_path(name: str) -> str:
    return f'../output/logs/ppo_compare_reward_settings/{name}'


def out_path(name: str) -> str:
    return f'../output/plots/ppo_compare_reward_settings/{name}'
```

```python
r_dense = load_training_results(log_path=log_path('dense'))
r_sparse = load_training_results(log_path=log_path('sparse'))
```

# Dense

```python
plot_maker_dense = PlotMakerPlaying(r_dense, folder=out_path('dense'), show=True)
plot_maker_dense.plot_training()
```

# Sparse

```python
plot_maker_sparse = PlotMakerPlaying(r_sparse, folder=out_path('sparse'), show=True)
plot_maker_sparse.plot_training()
```

# Comparison

```python
plot_maker_compare = PlotMakerPlaying(
    results={
        'Dense': r_dense,
        'Sparse': r_sparse,
    },
    folder=out_path('comparison'),
    show=True,
)
```

```python
plot_maker_compare.plot_eval('random')
```
