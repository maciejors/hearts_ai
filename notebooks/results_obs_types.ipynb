```python
from hearts_ai.rl.results import (
    load_training_results,
    PlotMakerPlaying,
    print_best_eval_run,
)
```

```python
def log_path(name: str) -> str:
    return f'../output/logs/obs_types/{name}'


def out_path(name: str) -> str:
    return f'../output/plots/obs_types/{name}'
```

```python
r_mctsrl_obs_full = load_training_results(log_path=log_path('mctsrl_obs_full'))
r_mctsrl_obs_compact = load_training_results(log_path=log_path('mctsrl_obs_compact'))

r_ppo_obs_full = load_training_results(log_path=log_path('../ppo_compare_training_variants/both_20'))
r_ppo_obs_compact = load_training_results(log_path=log_path('ppo_obs_compact'))
```

# MCTS+RL play & PPO card pass

```python
plot_maker_mctsrl = PlotMakerPlaying(
    {
        'Full obs.': r_mctsrl_obs_full,
        'Compact obs.': r_mctsrl_obs_compact,
    },
    folder=out_path('mctsrl_obs_compare'),
    show=True
)
```

```python
plot_maker_mctsrl.plot_eval('random')
```

```python
plot_maker_mctsrl.plot_eval('rule_based')
```
