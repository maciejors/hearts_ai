```python
import sys
sys.path.append('..')
```

```python
from hearts_ai.rl.results import (
    load_training_results,
    PlotMakerPlaying,
    print_best_eval_run,
)
```

```python
def log_path(name: str) -> str:
    return f'../output/logs/ppo_compare_training_variants/{name}'


def out_path(name: str) -> str:
    return f'../output/plots/ppo_compare_training_variants/{name}'
```

```python
r_separate = load_training_results(log_path=log_path('separate'))
r_both_20 = load_training_results(log_path=log_path('both_20'))
r_both_50 = load_training_results(log_path=log_path('both_50'))
```

# Training of 'separate' variant

```python
plot_maker_separate = PlotMakerPlaying(r_separate, folder=out_path('separate'), show=True)
plot_maker_separate.plot_training()
```

# Training of 'both' variant (more frequent swaps)

```python
plot_maker_both_20 = PlotMakerPlaying(r_both_20, folder=out_path('both_20'), show=True)
plot_maker_both_20.plot_training()
```

# Training of 'both' variant (less frequent swaps)

```python
plot_maker_both_50 = PlotMakerPlaying(r_both_50, folder=out_path('both_50'), show=True)
plot_maker_both_50.plot_training()
```

# Comparison

```python
plot_maker_compare = PlotMakerPlaying(
    results={
        'Separate': r_separate,
        'Continuous (swapped every 20 play PPO updates)': r_both_20,
        'Continuous (swapped every 50 play PPO updates)': r_both_50,
    },
    folder=out_path('comparison'),
    show=True,
)
```

```python
plot_maker_compare.plot_eval('random')
```

```python
plot_maker_compare.plot_eval('rule_based')
```
